```python
import os
import shutil

# Paths to directories
hr_dir = 'high_psnr_hr'
lc_dir = 'land_cover_final'
output_lc_dir = 'high_psnr_lc'

# Create the output directory if it doesn't exist
if not os.path.exists(output_lc_dir):
    os.makedirs(output_lc_dir)

# Get the list of HR images
hr_files = [f for f in os.listdir(hr_dir) if f.endswith('_rgb.png')]

# Extract the corresponding LC files
for hr_file in hr_files:
    base_name = hr_file.replace('_rgb.png', '')
    lc_file = f"{base_name}.tif"
    
    lc_image_path = os.path.join(lc_dir, lc_file)
    
    if os.path.exists(lc_image_path):
        shutil.copy(lc_image_path, os.path.join(output_lc_dir, lc_file))
        print(f"Copied: {lc_file}")
    else:
        print(f"Land cover image not found for: {hr_file}")

print(f"All corresponding land cover images have been copied to '{output_lc_dir}'.")
```
